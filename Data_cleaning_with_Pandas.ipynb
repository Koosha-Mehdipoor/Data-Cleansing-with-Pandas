we can see that because the strings such as g, mg and ... we cannot convert the values in float or int so we need to clean it.
- what we need to do is removing the measures and add them to the column name.
- we need to use the replace method and find the pattern in each value because it is not the same.
- for finding the regex patterns you can go to the https://regex101.com/
- in the website mentioned above we figured out that the pattern is [a-zA-Z] which works perfectly but we don't need it for now because for now we need to remove the numbers and just keep the measure to be able to add to column label for removing the numbers we can use this [^a-zA-Z]



```python
df.replace('[^a-zA-Z]','',regex = True)
```

We can see that our pattern didn't affect our column calories becuase it was not string and it was defined as int what can help is first changing all the data types to string.


```python
units = df.astype(str).replace('[^a-zA-Z]','',regex = True)
```

Now we have all the units, we need to know which values has the most occurance for each column and keep that value and Mode can help us here.

```python
units = units.mode()
units.head()
```

```python
for k in units:
  print(units[k].name + '_' + units[k].iloc[0])
```

```python
{units[k].name: units[k].name + '_' + units[k].iloc[0] for k in units}
```

we have created the new column names to feed the rename function but we can see that there is an issue with the columns which does not have any units so first we need to remove them.

```python
import numpy as np
units = units.replace('', np.nan).dropna(axis=1)
```

```python
mapper = {units[k].name: units[k].name + '_' + units[k].iloc[0] for k in units}
```

Now the issue is fixed, what we need to do now is using the rename function to rename all column names.

```python
df.rename(mapper = mapper,axis = 1,inplace = True)
```

```python
df.head()
```

Now we can see that our column names are changed and what we need to do is cleaning the values.

```python
df = df.astype(str).replace('[a-zA-Z]','',regex = True)
df.head()
```

As we can see by just removing the '^' from the regex pattern we have removed the non numeric part from the values, but still we have the issue of data type because they are all now object even if we see them as numbers and here the astype function can help us.

```python
df.replace('', np.nan,inplace = True)
```

```python
df = df.astype(float)
```
